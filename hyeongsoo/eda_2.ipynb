# Augmentation 해보기

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import albumentations as A
import cv2
```

## 좌우반전

```python
# 원본 이미지
image = cv2.imread("./data/train/00ec594b7cc89183.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image), plt.axis("off")
plt.show()
```

```python
transform_1 = A.Compose([
    A.HorizontalFlip(p=1)
])
```

```python
transformed_1 = transform_1(image=image)
```

```python
transformed_image_1 = transformed_1["image"]
```

```python
plt.imshow(transformed_image_1), plt.axis("off")
plt.show()
```

## 로테이션
- 시계반향으로 45도 회전(-45)
- 빈 부분은 흰색으로 채우기 (border_mode=cv2.BORDER_CONSTANT, value=[255,255,255])

```python
transform_2 = A.Compose([
    A.Rotate(limit=[-45,-45], border_mode=cv2.BORDER_CONSTANT, value=[255, 255, 255], p=1)
])
```

```python
transformed_2 = transform_2(image=image)
```

```python
transformed_image_2 = transformed_2["image"]
```

```python
plt.imshow(transformed_image_2), plt.axis("off")
plt.show()
```

## 블러(Blur)

### Blur (blur_limit = (4,4))

```python
transform_3 = A.Compose([
    A.Blur(blur_limit=(4,4), always_apply=True)
])

transformed_3 = transform_3(image=image)
transformed_image_3 = transformed_3["image"]

plt.imshow(transformed_image_3), plt.axis("off")
plt.show()
```

### Blur (blur_limit = (3,3))

```python
transform_3_2 = A.Compose([
    A.Blur(blur_limit=(3,3), always_apply=True)
])

transformed_3_2 = transform_3_2(image=image)

transformed_image_3_2 = transformed_3_2["image"]

plt.imshow(transformed_image_3_2), plt.axis("off")
plt.show()
```

## Gaussian Noise 추가
