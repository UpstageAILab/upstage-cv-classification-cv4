# Augmentation 해보기

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import albumentations as A
import cv2
```

## 좌우반전

```python
# 원본 이미지
image = cv2.imread("./data/train/fda9f9b14d231899.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image), plt.axis("off")
plt.show()
```

```python
transform_1 = A.Compose([
    A.HorizontalFlip(p=1)
])
```

```python
transformed_1 = transform_1(image=image)
```

```python
transformed_image_1 = transformed_1["image"]
```

```python
plt.imshow(transformed_image_1), plt.axis("off")
plt.show()
```

## 로테이션
- 시계반향으로 45도 회전(-45)
- 빈 부분은 흰색으로 채우기 (border_mode=cv2.BORDER_CONSTANT, value=[255,255,255])

```python
transform_2 = A.Compose([
    A.Rotate(limit=[-45,-45], border_mode=cv2.BORDER_CONSTANT, value=[255, 255, 255], p=1)
])
```

```python
transformed_2 = transform_2(image=image)
```

```python
transformed_image_2 = transformed_2["image"]
```

```python
plt.imshow(transformed_image_2), plt.axis("off")
plt.show()
```

## 블러(Blur)

```python
transform_3 = A.Compose([
    A.Blur(blur_limit=[4,4], always_apply=True)
])
```

```python
transformed_3 = transform_3(image=image)
```

```python
transformed_image_3 = transformed_3["image"]
```

```python
plt.imshow(transformed_image_3), plt.axis("off")
plt.show()
```

## Gaussian Noise 추가

```python
transform_4 = A.Compose([
    A.GaussNoise(var_limit=(300, 300), p=1.0)
])
```

```python
transformed_4 = transform_4(image=image)
```

```python
transformed_image_4 = transformed_4["image"]
```

```python
plt.imshow(transformed_image_4), plt.axis("off")
plt.show()
```

## ISONoise

```python
transform_5 = A.Compose([
    A.ISONoise(intensity=(1.5, 1.5), p=1.0)
])
```

```python
transformed_5 = transform_5(image=image)
```

```python
transformed_image_5 = transformed_5["image"]
```

```python
plt.imshow(transformed_image_5), plt.axis("off")
plt.show()
```
